```python
nvda["Return"] = nvda["Close"].pct_change()
```

```python
nvda["Volatility"] = nvda["Return"].rolling(window=10).std()
```

```python
horizons = [2, 5, 60, 250, 1000]

for horizon in horizons:
    rolling_avg = nvda["Close"].rolling(horizon).mean()
    nvda[f"Close_Ratio_{horizon}"] = nvda["Close"] / rolling_avg
```

```python
from xgboost import XGBClassifier
from sklearn.metrics import precision_score, accuracy_score
from sklearn.model_selection import train_test_split

# Define all predictors
predictors = [
    "Close", "Volume", "Open", "High", "Low",
    "Close_Ratio_2", "Close_Ratio_5", "Close_Ratio_60",
    "Close_Ratio_250", "Close_Ratio_1000",
    "Return", "Volatility"
]

# Split dataset: last 100 days for testing
train = nvda.iloc[:-100]
test = nvda.iloc[-100:]

# Initialize XGBoost model
model = XGBClassifier(eval_metric='logloss')
model.fit(train[predictors], train["Target"])

# Make predictions on the test set
predictions = model.predict(test[predictors])
```

```python
def backtest(data, model, predictors, start=2500, step=250, threshold=0.6):
    all_predictions = []

    for i in range(start, data.shape[0], step):
        train = data.iloc[0:i].copy()
        test = data.iloc[i:(i + step)].copy()

        # Use custom predict function with threshold
        combined = predict(train, test, predictors, model, threshold)
        all_predictions.append(combined)

    return pd.concat(all_predictions)
```

```python
def predict(train, test, predictors, model, threshold=0.6):
    model.fit(train[predictors], train["Target"])

    # Get probabilities of class 1 (stock going up)
    probs = model.predict_proba(test[predictors])[:, 1]

    # Apply threshold
    preds = (probs >= threshold).astype(int)

    # Return combined DataFrame with predictions and actuals
    return pd.DataFrame({
        "Predictions": preds,
        "Actual": test["Target"]
    }, index=test.index)
```

```python
# Run it
predictions = backtest(nvda, model, predictors, threshold=0.6)

# Accuracy
accuracy = (predictions["Predictions"] == predictions["Actual"]).mean()
print("Backtest Accuracy with Threshold 0.6:", round(accuracy, 3))
```

```python
from sklearn.metrics import precision_score, recall_score, confusion_matrix

# Accuracy
accuracy = (predictions["Predictions"] == predictions["Actual"]).mean()

# Precision
precision = precision_score(predictions["Actual"], predictions["Predictions"])

# Recall
recall = recall_score(predictions["Actual"], predictions["Predictions"])

# Confusion Matrix
cm = confusion_matrix(predictions["Actual"], predictions["Predictions"])

print("Backtest Evaluation Metrics:")
print("----------------------------")
print(f"Accuracy : {round(accuracy, 3)}")
print(f"Precision: {round(precision, 3)}")
print(f"Recall   : {round(recall, 3)}")
print("\nConfusion Matrix:")
print(cm)
```

```python
import matplotlib.pyplot as plt

# Create a rolling accuracy column
predictions["Correct"] = (predictions["Predictions"] == predictions["Actual"]).astype(int)
rolling_accuracy = predictions["Correct"].rolling(100).mean()

# Plot it
plt.figure(figsize=(12, 6))
plt.plot(rolling_accuracy, label="Rolling Accuracy (100 days)")
plt.title("Model Accuracy Over Time")
plt.xlabel("Date")
plt.ylabel("Accuracy")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
